```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_us_2021=pd.read_csv("US_2021.csv")
df_ca_2021=pd.read_csv("CA_2021.csv")
df_gb_2021=pd.read_csv("GB_2021.csv")
```

```python
df_us1_2021=df_us_2021.iloc[:,[1]]
df_ca1_2021=df_ca_2021.iloc[:,[1]]
df_gb1_2021=df_gb_2021.iloc[:,[1]]
```

# Unique hashtags count 2021

```python
import re
dict_us_21={}
hashtag_us_21=[]
for i in range(0, len(df_us1_2021)):
  k= df_us1_2021.at[i,"text"]
  ind = [i.start() for i in re.finditer("#", k)]
  for j in range(len(ind)):
    ind1=k[ind[j]:].find(" ")
    if(ind1==-1):
      word=k[ind[j]:]
    else:
      word=k[ind[j]:ind1]
    if(word not in dict_us_21):
      hashtag_us_21.append(word)
      dict_us_21[word]=1
print(hashtag_us_21)
print("number of unique hashtags in US 2021:" +str(len(hashtag_us_21)-1))
```

```python

dict_ca_21={}
hashtag_ca_21=[]
for i in range(0, len(df_ca1_2021)):
  k= df_ca1_2021.at[i,"text"]
  ind = [i.start() for i in re.finditer("#", k)]
  for j in range(len(ind)):
    ind1=k[ind[j]:].find(" ")
    if(ind1==-1):
      word=k[ind[j]:]
    else:
      word=k[ind[j]:ind1]
    if(word not in dict_ca_21):
      hashtag_ca_21.append(word)
      dict_ca_21[word]=1
print(hashtag_ca_21)
print("number of unique hashtags in CA 2021:" +str(len(hashtag_ca_21)-1))
```

```python

dict_gb_21={}
hashtag_gb_21=[]
for i in range(0, len(df_gb1_2021)):
  k= df_gb1_2021.at[i,"text"]
  ind = [i.start() for i in re.finditer("#", k)]
  for j in range(len(ind)):
    ind1=k[ind[j]:].find(" ")
    if(ind1==-1):
      word=k[ind[j]:]
    else:
      word=k[ind[j]:ind1]
    if(word not in dict_gb_21):
      hashtag_gb_21.append(word)
      dict_gb_21[word]=1
print(hashtag_gb_21)
print("number of unique hashtags in GB 2021:" +str(len(hashtag_gb_21)-1))
```

# **Tweet Time Analysis**

```python
df_us2_2021=df_us_2021.iloc[:,[3]]
df_ca2_2021=df_ca_2021.iloc[:,[3]]
df_gb2_2021=df_gb_2021.iloc[:,[3]]
```

```python
#dropping null values
df_us2_2021.dropna(inplace = True)
df_ca2_2021.dropna(inplace = True)
df_gb2_2021.dropna(inplace = True)
```

```python
def time_interval(c):
  h=(15*c)//60
  m=(15*c)%60
  if(m!=45):
    return str(h)+":"+str(m)+" - "+str(h)+":"+str(m+15)
  else:
    return str(h)+":"+str(m)+" - "+str(h+1)+":"+str(00)
```

```python
hour_us_21=[0 for i in range(96)]
maxcount=0
time=0
for i in range(0, len(df_us2_2021)):
  numh= int(df_us2_2021.at[i,"created_at"][11:13])
  numm= int(df_us2_2021.at[i,"created_at"][14:16])
  hour_us_21[((numh*60)+numm)//15-1]=hour_us_21[((numh*60)+numm)//15-1]+1
  if(hour_us_21[((numh*60)+numm)//15-1]>maxcount):
    time=((numh*60)+numm)//15
    maxcount=hour_us_21[((numh*60)+numm)//15-1]
print("maximum number of tweets in US 2021 took place between: "+time_interval(time))
```

```python
hour_ca_21=[0 for i in range(96)]
maxcount=0
time=0
for i in range(0, len(df_ca2_2021)):
  numh= int(df_ca2_2021.at[i,"created_at"][11:13])
  numm= int(df_ca2_2021.at[i,"created_at"][14:16])
  hour_ca_21[((numh*60)+numm)//15-1]=hour_ca_21[((numh*60)+numm)//15-1]+1
  if(hour_ca_21[((numh*60)+numm)//15-1]>maxcount):
    time=((numh*60)+numm)//15
    maxcount=hour_ca_21[((numh*60)+numm)//15-1]
print("maximum number of tweets in CA 2021 took place between: "+time_interval(time))
```

```python
hour_gb_21=[0 for i in range(96)]
maxcount=0
time=0
for i in range(0, len(df_gb2_2021)):
  numh= int(df_gb2_2021.at[i,"created_at"][11:13])
  numm= int(df_gb2_2021.at[i,"created_at"][14:16])
  hour_gb_21[((numh*60)+numm)//15-1]=hour_gb_21[((numh*60)+numm)//15-1]+1
  if(hour_gb_21[((numh*60)+numm)//15-1]>maxcount):
    time=((numh*60)+numm)//15
    maxcount=hour_gb_21[((numh*60)+numm)//15-1]
print("maximum number of tweets in GB 2021 took place between: "+time_interval(time))
```
